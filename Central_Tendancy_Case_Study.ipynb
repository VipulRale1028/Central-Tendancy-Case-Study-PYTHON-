# **Task 2 - View the distribution of the various features in the data set and calulate their central tendancy**

```python
# create histogram for "Fixed acidity"

plt.figure(figsize=(12, 5))

sns.histplot(data = data , x = "fixed acidity",color = "blue",
edgecolor = "linen",bins = 5)

plt.title("Histogram of Fixed Acidity")
plt.xlabel("Fixed Acidity")
plt.ylabel("count")

plt.show()
```

```python
# Now Calculate the Mean of "fixed acidity"

round(data['fixed acidity'].mean(),2)
```

```python
# Now Calculate the Median of "fixed acidity"

data["fixed acidity"].median()
```

```python
# create histogram for "Fixed acidity" and also show the mean and median

plt.figure(figsize=(12, 5))

sns.histplot(data = data , x = "fixed acidity",color = "orange",
edgecolor = "linen",bins = 5)
plt.title("Histogram of Fixed Acidity")
plt.xlabel("Fixed Acidity")
plt.ylabel("count")
plt.vlines(data["fixed acidity"].mean(),ymin = 0, ymax = 4000, color = "blue",label = "Mean")
plt.vlines(data["fixed acidity"].median(),ymin = 0, ymax = 4000, color = "red",label = "Median")
plt.legend()
plt.show()
```

```python
# create histogram for "Volatile acidity"

plt.figure(figsize=(12, 5))

sns.histplot(data = data , x = "volatile acidity",color = "green",
edgecolor = "linen",bins = 5)

plt.title("Histogram of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("count")

plt.show()
```

```python
# Plot using distplot for "Volatile Acidity"

plt.figure(figsize = (10,5))

sns.distplot(data["volatile acidity"],color = "yellow")
plt.title("Distribution of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Density")
plt.show()
```

```python
# Calculate Skewness in "Volatile Acidity"

a = round(data["volatile acidity"].skew(),2)

print("Skewness in Volatile Acidity is : ",a)

print("Hence....!  It is Positive Skew")
```

```python
# Mean of "Volatile Acidity"

m = round(data["volatile acidity"].mean(),2)

print("Mean of the volatile acidity is : ",m)
```

```python
# Median of "Volatile Acidity"

n = round(data["volatile acidity"].median(),2)

print("Median of the volatile acidity is : ",n)
```

```python
# create histogram for "Volatile acidity" and also show the mean and median

plt.figure(figsize=(12, 5))

sns.histplot(data = data , x = "volatile acidity",color = "yellow",
edgecolor = "linen",bins = 5)
plt.title("Histogram of Volatile Acidity")
plt.xlabel("Volatile Acidity")
plt.ylabel("Density")
plt.vlines(data["volatile acidity"].mean(),ymin = 0, ymax = 4000, color = "blue",label = "Mean")
plt.vlines(data["volatile acidity"].median(),ymin = 0, ymax = 4000, color = "red",label = "Median")
plt.legend()
plt.show()
```

```python
# create histogram for "Citric acidity"

plt.figure(figsize=(12, 5))

sns.histplot(data = data , x = "citric acid",color = "blue",
edgecolor = "linen",bins = 5)
plt.title("Histogram of Citric Acid")
plt.xlabel("Citric Acid")
plt.ylabel("count")
plt.legend()
plt.show()
```

```python
# Mean of "Citric Acid"

x = round(data["citric acid"].mean(),2)

print("Mean of the citric acid is : ",x)
```

```python
# Median of "Citric Acid"

z = round(data["citric acid"].median(),2)

print("Median of the citric acid is : ",z)
```

```python
# CountPlot of the "Quality" feature

plt.figure(figsize=(11,6))

sns.countplot(data["quality"])
plt.title("Count Plot of Quality")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.show()
```

```python
# Create a new Pandas Series called "rep_acid" that contains the details of the representative quality for the different types of acids

rep_acid = pd.DataFrame(index = ['fixed acidity','volatile acidity','citric acid','quality'],
                     data = [data['fixed acidity'].mean(),data['volatile acidity'].mean(),
                             data['citric acid'].mean(),data['quality'].value_counts().index[0]])
```
